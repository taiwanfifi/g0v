將書籍依年齡分類，包括"嬰幼兒及學齡前"、"圖畫書4-8"、"兒童小說9-12"、"青少年小說13-"。  
http://kids.tpml.edu.tw/sp.asp?xdurl=superXD/newbook.asp&mp=100

```python
import requests 
import bs4
import time
import pandas as pd
import re
```

### 每本書的資訊

```python
def get_book_detail(book_link, bookDic):
    
    sub_link = 'http://kids.tpml.edu.tw/'+ book_link
    response = requests.get(sub_link)    
    soup = bs4.BeautifulSoup(response.text,"html.parser")   
    
    book_content = soup.find('ul','aboutbook') 
    book_content = str(book_content)

    jojo = re.split('<em>|</em>|</li>|<li>', book_content)
    #print(len(jojo))
    #for i,j in enumerate(jojo):
    #    print(i, j)

    a = b = c = 0        
    for i in range(len(jojo)):
        if '書名：' in jojo[i] and '叢' not in jojo[i]:
            print('  -> {}{}'.format(jojo[i], jojo[i+1]))  
            bookDic['title'].append(jojo[i+1])
            a = 1
        if '適讀年齡' in jojo[i]:
            print('  -> {}{}'.format(jojo[i], jojo[i+1]))  
            bookDic['age'].append(jojo[i+1])
            b = 1
        if '主題詞' in jojo[i]:
            lis = []
            print('  -> {}{}'.format(jojo[i], jojo[i+1]))  
            topic = jojo[i+1].split(';')
            for t in topic:
                lis.append(t.strip())
            bookDic['topic'].append(lis)
            c = 1
    if a==0 and b==0 and c==0:  #代表都沒抓到，直接跳過
        pass
    else: #其一沒抓到，新增None
        if a==0:bookDic['title'].append('None') 
        if b==0:bookDic['age'].append('None') 
        if c==0:bookDic['topic'].append('None')
        
    return bookDic    

### 測試區 ###    
#bookDic = {
#    'title':[],     #書名 titles    
#    'age':[],             #適讀年齡 ages
#    'topic':[]             #主題詞 list topics
#}
#book_link = "sp.asp?xdurl=bookDetail.asp&mp=100&gicuitem=177904" 
#get_book_detail(book_link, bookDic)
```

### 獲取每一頁中的單書連結

```python
def get_page_titles(link, bookDic):
   
    response = requests.get(link)   
    soup = bs4.BeautifulSoup(response.text,"html.parser")    
    
    title = {}
    books = soup.find_all('a')
    for book in books:
        if '/sp.asp?xdurl=bookDetail.asp&mp=100&gicui' in book['href'] and book.text!='':
        
            book_title = book.text
            book_link = book['href']
            print('=======')
            print('書標題:{}'.format(book_title))
            print('書連結:{}'.format(book_link))  
            get_book_detail(book_link, bookDic)
            
    return bookDic
```

```python
bookDic = {
    'title':[],     #書名 titles
    'age':[],             #適讀年齡 ages
    'topic':[]             #主題詞 list topics
}

for i in range(23):
    i = i+1
    print('第{}頁'.format(i))
    link = "http://kids.tpml.edu.tw/sp.asp?xdurl=superXD/newbook.asp&nowPage="+ str(i) + "&mp=100&cat="   # 組成 正確 URL
    get_page_titles(link, bookDic)  # 執行單頁面網頁爬蟲 
```
